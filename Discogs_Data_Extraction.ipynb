```python
# List for storing all information regarding collection release_ids
all_releases_collection = []

# Loop through all collection release_ids
for release_id in collection_release_ids:
    release = d.release(release_id)
    time.sleep(1)  # short break for rate limit
   
    formats_info = release.formats
    formats = [] # List for storing all information regarding formats

    # Loop through the formats of release_ids
    for format_entry in formats_info:
        format_name = format_entry.get('name', 'Unknown Format')
        descriptions = ', '.join(format_entry.get('descriptions', []))
        #descriptions = format_entry.get('descriptions', [])[0] if format_entry.get('descriptions') else '' #only the first element in the list

        # String for format
        formats.append(f"{format_name} ({descriptions})")
        

    # Code if you want the urls from all images and not only the very first one
    #image_urls = [image['resource_url'] for image in release.images] 

    # Get the very first image_url of the images
    #first_image_uri = release.images[0]['uri'] if release.images else None
    first_image_url = release.images[0]['resource_url'] if release.images else None

    # Get all information regarding the release_ids
    r1 = {
            'release_id': release.id,
            'title': release.title,
            'artist': ', '.join([artist.name for artist in release.artists]),
            #'artists_sort' : release.artists_sort,
            'label': ', '.join([label.name for label in release.labels]),
            'year': release.year,
            'country': release.country,
            'genres': ', '.join(release.genres),
            'styles': ', '.join(release.styles),
            'tracklist': ', '.join([track.title for track in release.tracklist]),
            'formats': '; '.join(formats),
            #'formats2':''.join(formats),
            'url' : release.url,
            #'image': ', '.join(image_uris), 
            'first_image_uri': first_image_url,  
        }

    # Convert to dataframe
    df1 = pd.DataFrame(r1.items()) 

    # Transpose columns/rows
    df1 = df1.T  
    # Use first row as headers
    df1.columns = df1.iloc[0] 
    # Remove first row
    df1 = df1[1:] 

    # add Datframe to list
    all_releases_collection.append(df1)

# Concat all dataframes
final_collection = pd.concat(all_releases_collection, ignore_index=True)
# Add Datetime of Data Extraction Date
final_collection['Update_Datetime']= date_time_now


final_collection
```

**Get Wantlist Data**

```python
# List for the release_ids of the personal wantlist
wantlist_release_ids = []

# Loop through the wantlist folder and extract all release_ids and store them into one list
for item in me.wantlist:
    release_id = item.release.id
    wantlist_release_ids.append(release_id)

print(wantlist_release_ids)
```

```python
# List for storing all information regarding release_ids
all_releases_wantlist = []

# Loop through all release_ids
for release_id in wantlist_release_ids:
    release = d.release(release_id)
    time.sleep(1)  
    
    formats_info = release.formats
    formats = [] # List for storing all information regarding formats

    # Loop through the formats of release_ids
    for format_entry in formats_info:
        format_name = format_entry.get('name', 'Unknown Format')
        descriptions = ', '.join(format_entry.get('descriptions', []))
    
        # String for format
        formats.append(f"{format_name} ({descriptions})")

    # Code if you want the urls from all images and not only the very first one
    #image_urls = [image['resource_url'] for image in release.images] 

    # Get the very first image_url of the images
    #first_image_uri = release.images[0]['uri'] if release.images else None
    first_image_url = release.images[0]['resource_url'] if release.images else None

    # Get all information regarding the release_ids
    r2 = {
            'release_id': release.id,
            'title': release.title,
            'artist': ', '.join([artist.name for artist in release.artists]),
            #'artist_sort' : release.artists_sort,
            'label': ', '.join([label.name for label in release.labels]),
            'year': release.year,
            'country': release.country,
            'genres': ', '.join(release.genres),
            'styles': ', '.join(release.styles),
            'tracklist': ', '.join([track.title for track in release.tracklist]),
            'formats': '; '.join(formats),
            'url' : release.url,
            #'image': ', '.join(image_uris), 
            'first_image_uri': first_image_url,  

        }
    
    # Convert to dataframe
    df2 = pd.DataFrame(r2.items()) 

    # Transpose columns/rows
    df2 = df2.T  
    # Use first row as headers
    df2.columns = df2.iloc[0] 
    # Remove first row
    df2 = df2[1:] 

    # add dataframe to list
    all_releases_wantlist.append(df2)

# Concat all dataframes
final_wantlist = pd.concat(all_releases_wantlist, ignore_index=True)
# Add Datetime of Data Extraction Date
final_wantlist['Update_Datetime']= date_time_now


final_wantlist
```
